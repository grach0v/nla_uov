```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg.lapack import dtrtri
```

```python
A = np.array([
    [4, 0.5, 1],
    [2, 0.5, 1],
    [4, 1, -0.9],
])

b = np.array([1, 1, 1], dtype=np.float64)
```

```python
L = np.tril(A)
L_inv = dtrtri(L, lower=0)[0]
L @ L_inv
```

Why `dtrtri` does not work?  
I will use `np.linalg.inv` for now, becouse matrix is small and it does not matter a lot.  
But for bigger systems special methods for inversion triangular matricies supposed to be used.

```python
def richardson_iter(A, b, x_0, dump=1, max_iter=100, eps=1e-6, save_hist=True):
    x = x_0.copy()
    
    if save_hist:
        hist_residual = []

    for _ in range(max_iter):
        res = b - A @ x
        res_norm = np.linalg.norm(res)

        if res_norm < eps:
            break

        if save_hist:
            hist_residual.append(res_norm)

        x += dump * res
    
    if save_hist:
        return x, hist_residual
    
    return x
```

```python
def jacobi_iter(A, b, x_0, max_iter=100, eps=1e-6, save_hist=True):
    x = x_0.copy()
    
    if save_hist:
        hist_residual = []

    D = np.diag(A) # Note D is vector, not matrix 
    L_U = A - D

    for _ in range(max_iter):
        res = b - A @ x
        res_norm = np.linalg.norm(res)

        if res_norm < eps:
            break

        if save_hist:
            hist_residual.append(res_norm)

        x = (1 / D) * (b - (L_U @ x))
    
    if save_hist:
        return x, hist_residual
    
    return x
```

```python
def gauss_siden_iter(A, b, x_0, max_iter=100, eps=1e-6, save_hist=True):
    x = x_0.copy()
    
    if save_hist:
        hist_residual = []

    L = np.tril(A, 0)
    U = np.triu(A, 1)
    L_inv = np.linalg.inv(L)

    for _ in range(max_iter):
        res = b - A @ x
        res_norm = np.linalg.norm(res)

        if res_norm < eps:
            break

        if save_hist:
            hist_residual.append(res_norm)

        x = L_inv @ (b - U @ x)

    if save_hist:
        return x, hist_residual
    
    return x
```

## Initial A

```python
iters = 100
_, hist_rich_1 = richardson_iter(A, b, b, 1, iters)
_, hist_rich_01 = richardson_iter(A, b, b, 0.1, iters)
_, hist_rich_001 = richardson_iter(A, b, b, 0.01, iters)
_, hist_jacobi = jacobi_iter(A, b, b, iters)
_, hist_gauss = gauss_siden_iter(A, b, b, iters)
x = np.arange(iters) + 1
```

```python
plt.plot(x, hist_rich_1, label='richardson dump=1')
plt.plot(x, hist_rich_01, label='richardson dump=0.1')
plt.plot(x, hist_rich_001, label='richardson dump=0.01')
plt.plot(x, hist_jacobi, label='jacobi')
plt.plot(x, hist_gauss, label='gauss_siden')

ax = plt.gca()
ax.set_ylim([0, 100])
plt.legend()
plt.show()
```
